```python
class WaterJugPuzzle():

  def __init__(self):
    self.FiveLiterJug = 0
    self.ThreeLiterJug = 0

  # check if 5 liter jug is overflowing or negative value
  def is_FiveLiterJug_out_of_bounds(self):
    if self.FiveLiterJug > 5 or self.FiveLiterJug < 0:
        return True
    else:
        return False

  # check if 3 liter jug is overflowing or negative value
  def is_ThreeLiterJug_out_of_bounds(self):
    if self.ThreeLiterJug > 3 or self.ThreeLiterJug < 0:
        return True
    else:
        return False

  # check if state is okay
  def state_ok(self):
    if self.is_FiveLiterJug_out_of_bounds() or self.is_ThreeLiterJug_out_of_bounds():
        return False
    else:
        return True

  # save state
  def save_state(self):
    self._FiveLiterJug = self.FiveLiterJug
    self._ThreeLiterJug = self.ThreeLiterJug

  # undo state
  def undo_state(self):
    self.FiveLiterJug = self._FiveLiterJug
    self.ThreeLiterJug = self._ThreeLiterJug

  ## Rule 1: Fill the 5 Liter Jug up all the way
  def fill_FiveLiterJug(self):
    while self.state_ok():
      self.save_state()
      self.FiveLiterJug += 1
      if not self.state_ok():
        self.undo_state()
        break

  ## Rule 2: Fill the 3 Liter Jug up all the way
  def fill_ThreeLiterJug(self):
    while self.state_ok():
      self.save_state()
      self.ThreeLiterJug += 1
      if not self.state_ok():
        self.undo_state()
        break

  ## Rule 3: Dump out 5 Liter Jug
  def dump_FiveLiterJug(self):
    self.save_state()
    self.FiveLiterJug = 0
    if not self.state_ok():
      self.undo_state()

  ## Rule 4: Dump out 3 Liter Jug
  def dump_ThreeLiterJug(self):
    self.save_state()
    self.ThreeLiterJug = 0
    if not self.state_ok():
      self.undo_state()

  ## Rule 5: Pour 5 Liter Jug into Three Liter Jug
  def pour_FiveLiterJug_into_ThreeLiterJug(self):
    while self.state_ok():
      self.save_state()
      self.FiveLiterJug -= 1
      self.ThreeLiterJug += 1
      if not self.state_ok():
        self.undo_state()
        break

  ## Rule 6: Pour 3 Liter Jug into 5 Liter Jug
  def pour_ThreeLiterJug_into_FiveLiterJug(self):
    while self.state_ok():
      self.save_state()
      self.ThreeLiterJug -= 1
      self.FiveLiterJug += 1
      if not self.state_ok():
        self.undo_state()
        break

  ## Display what is where:
  def print_state(self):
    print("===> Five Liter Jug: {f}, Three Liter Jug: {t}".format(f=self.FiveLiterJug, t=self.ThreeLiterJug))
```

```python
wjp = WaterJugPuzzle()

# print initial state
wjp.print_state()

print("-----------------------")
# fill 5 liter jug
wjp.fill_FiveLiterJug()
wjp.print_state()

print("-----------------------")
# pour 5 liter jug into 3 liter jug
wjp.pour_FiveLiterJug_into_ThreeLiterJug()
wjp.print_state()

print("-----------------------")
# empty the 3 liter jug
wjp.dump_ThreeLiterJug()
wjp.print_state()

print("-----------------------")
# pour 5 liter jug into 3 liter jug
wjp.pour_FiveLiterJug_into_ThreeLiterJug()
wjp.print_state()

print("-----------------------")
# fill 5 liter jug
wjp.fill_FiveLiterJug()
wjp.print_state()

print("-----------------------")
# pour 5 liter jug ino the 3 liter jug
wjp.pour_FiveLiterJug_into_ThreeLiterJug()
wjp.print_state()
```
